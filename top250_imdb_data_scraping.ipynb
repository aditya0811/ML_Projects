```
from bs4 import BeautifulSoup
```

```
import os
import pandas as pd
df_list=[]
```

```
with open('imdb.html') as file:
  soup=BeautifulSoup(file,"lxml")
```

```
with open('imdb.html') as file:
  soup=BeautifulSoup(file,"lxml")
  movie_lists= soup.find('tbody',class_="lister-list").find_all('tr')
  for movie_list in movie_lists:
    movie_name= movie_list.find('td',class_="titleColumn").find('a').contents[0]
    Rating= float(str(movie_list.find('td',class_="posterColumn").find_all('span')[1])[18:-19])
    users= int(str(movie_list.find('td',class_="posterColumn").find_all('span')[3])[18:-19])
    year = int(movie_list.find('td',class_='titleColumn').find('span').contents[0][1:-1])
    dire= str(movie_list.find('td',class_="titleColumn").contents[1]).split('title=')[1].split('>')[0].split(',')
    direc=dire[0][1:-7]
    Characters=str(dire[1:])[1:-3]+ "'"
  
    df_list.append({'title': movie_name,
                        'Year_of_Release': year,
                        'Imdb_Rating': Rating,
                        'Number_of_Users_rated':users,
                        'Director':direc,
                        'Main_Characters':Characters})
  

  
df = pd.DataFrame(df_list, columns = ['title', 'Year_of_Release', 'Imdb_Rating','Number_of_Users_rated','Director','Main_Characters'])
df
```
